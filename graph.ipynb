```python
import json

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
with open('errors.json') as file:
    df = pd.DataFrame(json.load(file), columns=['word', 'errors'])
```

```python
df.errors.median()
```

```python
ax = df.groupby('errors').agg(['count']).plot(kind='bar')
ax.get_legend().remove()
plt.xticks(rotation=0)
plt.axvline(5.5, color='red', linestyle='--')
plt.text(5.6, 11100 ,'HANGMAN IS DEAD', rotation=0, fontsize=18, color='red')
ax.set_ylim(0, 14000)
plt.xlabel('Number of wrong letter guesses before solving', fontsize=14)
plt.ylabel('Number of words', fontsize=14)
plt.show()
```
